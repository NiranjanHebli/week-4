## 3. Broadcasting

```python
# Add 1D array to 2D array
mat = np.array([[1, 2, 3],
                [4, 5, 6]])
vec = np.array([10, 20, 30])
```

```python
print(mat + vec)   # row-wise addition
```

```python
# Multiply matrix by scalar
print(mat * 2)
```

```python
# Multiply matrix by vector (column-wise broadcasting)
col_vec = np.array([1, 2]).reshape(2, 1)
print(mat * col_vec)
```

### Explanation of broadcasting:

Broadcasting in NumPy is a way of performing operations on arrays of different shapes by automatically expanding the smaller array to match the larger one. It works when dimensions are either equal or one of them is 1, allowing NumPy to replicate values along that axis. This makes element‑wise operations efficient without needing explicit loops.

NumPy automatically expands dimensions when shapes are compatible.

[2x3] + [3] -> [2x3] + [1x3] -> broadcast across rows.

[2x3] * scalar -> scalar is treated as [1x1], expanded to [2x3].

[2x3] * [2x1] -> broadcast column vector across columns.

## 4. Vectorised Operations

```python
arr = np.array([-3, -1, 0, 2, 4])
```

```python
# Square & cube
print(arr**2)
print(arr**3)
```

```python
# Replace negatives with 0
arr_nonneg = np.where(arr < 0, 0, arr)
print(arr_nonneg)
```

```python
# Normalize (scale between 0 and 1)
arr_norm = (arr - arr.min()) / (arr.max() - arr.min())
print(arr_norm)
```

## 5. Dataset Operations

```python
data = np.array([[12, 7, 9],
                 [4, 15, 6],
                 [10, 2, 18]])
```

```python
# Top 5 maximum values
top5 = np.sort(data.flatten())[-5:]
print(top5)
```

```python
# Row-wise & column-wise sums
print(np.sum(data, axis=1))  # row sums
print(np.sum(data, axis=0))  # column sums
```

```python
# Indices of values greater than threshold
threshold = 10
indices = np.argwhere(data > threshold)
print(indices)
```

## 3. Performance Comparison

```python
import time
```

```python
# Large array
arr = np.arange(1_000_000)
```

```python
# Python loop sum
start = time.time()
s = 0
for val in arr:
    s += val
end = time.time()
print("Loop sum:", s, "Time:", end - start)
```

```python
# NumPy sum
start = time.time()
s_np = np.sum(arr)
end = time.time()
print("NumPy sum:", s_np, "Time:", end - start)
```

The time difference between summing a large array with a Python loop versus using NumPy is significant because of how the computations are executed under the hood:

- Python loops run in the interpreter, processing one element at a time. Each iteration involves overhead (loop control, variable updates, type checks), which makes them slow for millions of elements.

- NumPy operations are vectorised: they delegate the heavy lifting to optimized C and Fortran libraries (like BLAS/LAPACK). These libraries operate directly on contiguous blocks of memory, using low-level machine instructions and sometimes parallelization.

As a result, NumPy avoids Python’s per‑element overhead and executes the entire operation in compiled code, which is orders of magnitude faster.
