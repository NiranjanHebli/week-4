```python
# Student class with name, marks, and grade calculation
class Student:
    # Constructor to initialize name and marks
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    # Method to calculate grade based on marks
    def calculate_grade(self):
        if self.marks >= 80:
            return "A"
        elif self.marks >= 60:
            return "B"
        else:
            return "C"

    def __str__(self):
        return f"Name: {self.name}, Marks: {self.marks}, Grade: {self.calculate_grade()}"
```

```python
s1 = Student("Rahul", 85)
s2 = Student("Anita", 72)
s3 = Student("Kiran", 55)

print(s1)
print(s2)
print(s3)
```
