## 1. Using only loops (no built-in functions)

### Find maximum and minimum element in a list

```python
# get min and max from a list of numbers
def min_max(numbers):
    """ Find the minimum and maximum numbers in a list."""
    max_num = numbers[0]
    min_num = numbers[0]

    for num in numbers:
        if num > max_num:
            max_num = num
        if num < min_num:
            min_num = num
    return min_num, max_num
```

```python
numbers = [12, 45, 7, 89, 23]
min_num, max_num = min_max(numbers)
print("Minimum:", min_num)
print("Maximum:", max_num)
```

### Count frequency of each element using a dictionary

```python
# count the frequency of each number in a list of numbers by creating a dictionary
def count_frequencies(numbers):
    """ Count the frequency of each number in a list."""
    freq = {}
    for num in numbers:
        if num in freq:
            freq[num] += 1
        else:
            freq[num] = 1
    return freq
```

```python
numbers = [12, 45, 7, 89, 23, 12, 45]
freq = count_frequencies(numbers)
print("Frequencies:", freq)
```

## 2. Using a while loop

### Reverse a number

```python
# reverse a number by extracting its digits and constructing the reversed number
def reverse_number(num):
    """ Reverse a number by extracting its digits."""
    reverse = 0
    while num > 0:
        digit = num % 10
        reverse = reverse * 10 + digit
        num = num // 10

    return reverse
```

```python
num = 123
reverse = reverse_number(num)
print("Reversed number:", reverse)
```

### Check whether a number is palindrome

```python
# check whether the given number is a palindrome by comparing the original number with the reversed number
def is_palindrome(num):
    """ Check if a number is a palindrome."""
    original_num = num
    reverse = reverse_number(num)
    if original_num == reverse:
        print(f"{original_num} is a palindrome.")
    else:
        print(f"{original_num} is not a palindrome.")
```

## 3. Work with tuples and dictionaries

### Convert list of tuples into dictionary without built-in conversion

```python
# convert a list of pairs into a dictionary by iterating through the list and adding each pair to the dictionary
def pairs_to_dict(pairs):
    """ Convert a list of pairs into a dictionary."""
    result = {} 
    for item in pairs:
        key = item[0]
        value = item[1]
        result[key] = value

    return result
```

```python
pairs = [('a', 10), ('b', 20), ('c', 30)]
result = pairs_to_dict(pairs)
print("Resulting dictionary:", result)
```

### Find key with highest value using loops only

```python

# find the key with the highest value in a dictionary by iterating through the dictionary and keeping track of the maximum
def find_highest_value_key(data):
    """ Find the key with the highest value in a dictionary."""
    max_key = None
    max_value = None
    
    for key in data:
        if max_value is None or data[key] > max_value:
            max_value = data[key]
            max_key = key
    return max_key
```

```python
data = {'A': 50, 'B': 80, 'C': 65}
highest_key = find_highest_value_key(data)
print("Key with the highest value:", highest_key)
```

## 4. Function using *args

### Return sum and average without using sum()

```python
# calculate the sum and average of a list of numbers by iterating through the list and keeping track of the total and count
def calculate(*args):
    """ Calculate the sum and average of a list of numbers."""
    total = 0
    count = 0

    for num in args:
        total += num
        count += 1

    average = total / count
    return total, average
```

```python
total, average = calculate(10, 20, 30, 40)
print("Total:", total)
print("Average:", average)
```

## 5. Function using **kwargs

### Return student with highest marks

```python
# find the student with the highest marks by iterating through a dictionary of student names and marks.
def top_student(**kwargs):
    """ Find the student with the highest marks."""
    highest_name = None
    highest_marks = None

    for name in kwargs:
        if highest_marks is None or kwargs[name] > highest_marks:
            highest_marks = kwargs[name]
            highest_name = name

    return highest_name, highest_marks
```

```python
name, marks = top_student(Niranjan=85, Om=92, Hiten=88)
print(f"Top student: {name} with marks {marks}")
```

# Part B - Stretch Problem

```python
# create a class Vector that represents a mathematical vector and implement methods for vector addition, subtraction, scalar multiplication, and dot product.
class Vector:
    def __init__(self, values):
        self.values = values

    # Vector addition
    def __add__(self, other):
        result = []
        for i in range(len(self.values)):
            result.append(self.values[i] + other.values[i])
        return Vector(tuple(result))

    # Vector subtraction
    def __sub__(self, other):
        result = []
        for i in range(len(self.values)):
            result.append(self.values[i] - other.values[i])
        return Vector(tuple(result))

    # Scalar multiplication
    def __mul__(self, scalar):
        result = []
        for i in range(len(self.values)):
            result.append(self.values[i] * scalar)
        return Vector(tuple(result))
    
    # Dot product
    def dot(self, other):
        result = 0
        for i in range(len(self.values)):
            result += self.values[i] * other.values[i]
        return result

    def __repr__(self):
        return f"Vector{self.values}"
```

```python
v1 = Vector((1, 2, 3))
v2 = Vector((4, 5, 6))

print("Vector 1:", v1)
print("Vector 2:", v2)

# Addition
print("Addition:", v1 + v2)

# Subtraction
print("Subtraction:", v1 - v2)

# Scalar multiplication
print("Scalar Multiplication:", v1 * 3)

# Dot product
print("Dot Product:", v1.dot(v2))

v3 = Vector((10, 20))
v4 = Vector((5, 15))
```
