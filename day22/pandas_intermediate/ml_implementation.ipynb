# Part A :- Concept Application

Below is a complete example using the Titanic dataset dataset (CSV has 800+ rows, so it satisfies the 500+ rows requirement).

## 1. Identify ML Problem Type

This dataset is Supervised Learning because:

Input features are already available (Age, Fare, Sex, Pclass, etc.)

Target values are known

### Why Supervised?

The model learns from labeled examples:

For classification -> target = Survived

For regression -> target = Fare

### Classification or Regression?

Classification: Predict whether a passenger survived (0 or 1)

Regression: Predict fare paid (continuous numeric value)

## 2. Data Handling with Pandas

```python
import pandas as pd

# Load dataset
df = pd.read_csv("titanic.csv")

# View first rows
print(df.head())

# Check missing values
print(df.isnull().sum())

# Fill missing Age values with mean
df['Age'] = df['Age'].fillna(df['Age'].mean())

# Drop rows with missing Embarked
df = df.dropna()

# Select relevant columns
df = df[['Age', 'Fare', 'Pclass', 'Sex', 'Survived']]

# Convert categorical Sex into numeric
df['Sex'] = df['Sex'].map({'male':0, 'female':1})

print(df.head())
```

## 3. Regression Task

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

X = df[['Age', 'Pclass', 'Sex']]
y = df['Fare']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LinearRegression()
model.fit(X_train, y_train)

pred = model.predict(X_test)

mse = mean_squared_error(y_test, pred)

print("MSE:", mse)
```

## 4. Classification Task

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

X = df[['Age', 'Fare', 'Pclass', 'Sex']]
y = df['Survived']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

clf = LogisticRegression()
clf.fit(X_train, y_train)

pred = clf.predict(X_test)

acc = accuracy_score(y_test, pred)

print("Accuracy:", acc)
```

## 5. Comparison: Regression vs Classification



Regression:

- Predict continuous numeric values (Fare)

- Evaluation metric: Mean Squared Error (MSE)

Classification:

- Predict categorical values (Survived)

- Evaluation metric: Accuracy Score
