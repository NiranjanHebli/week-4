```python
import pandas as pd
from sklearn.datasets import load_iris

# Load Iris dataset
iris = load_iris(as_frame=True)
df = iris.frame

# Filter rows where sepal length is greater than average
average_value = df['sepal length (cm)'].mean()

filtered_rows = df[df['sepal length (cm)'] > average_value]

print("Average sepal length:", average_value)
print(filtered_rows.head())
```
