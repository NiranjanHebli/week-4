# Part A - Concept Application

```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
import math
```

### 1. Synthetic dataset from normal distribution

```python
np.random.seed(42)

# Generate dataset
data = np.random.normal(loc=50, scale=10, size=1000)

# Compute statistics
mean_val = np.mean(data)
variance_val = np.var(data)
std_val = np.std(data)

print("Normal Distribution Dataset")
print("Mean:", mean_val)
print("Variance:", variance_val)
print("Standard Deviation:", std_val)

# Histogram
plt.hist(data, bins=30)
plt.title("Histogram of Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

print("Observation: Histogram shows bell shaped symmetric distribution.")
```

### 2. PMF for Bernoulli Distribution

```python
def bernoulli_pmf(x, p):
    if x == 1:
        return p
    elif x == 0:
        return 1 - p
    else:
        return 0

p = 0.6

pmf_0 = bernoulli_pmf(0, p)
pmf_1 = bernoulli_pmf(1, p)

print("Bernoulli PMF")
print("P(X=0):", pmf_0)
print("P(X=1):", pmf_1)
print("Total Probability:", pmf_0 + pmf_1)
```

### 3. Manual PDF of Normal Distribution

```python
def normal_pdf(x, mu, sigma):
    coeff = 1 / (sigma * math.sqrt(2 * math.pi))
    exponent = math.exp(-((x - mu) ** 2) / (2 * sigma ** 2))
    return coeff * exponent

x = 50
pdf_value = normal_pdf(x, 50, 10)

print("Manual Normal PDF")
print("PDF at x=50:", pdf_value)
```

 ### 4. Mean, Median, Mode, Skewness

```python
mean_calc = np.mean(data)
median_calc = np.median(data)

# Mode approximation
rounded_data = np.round(data)
mode_calc = Counter(rounded_data).most_common(1)[0][0]

# Skewness
skewness = 3 * (mean_calc - median_calc) / std_val

print("Descriptive Statistics")
print("Mean:", mean_calc)
print("Median:", median_calc)
print("Mode:", mode_calc)
print("Skewness:", skewness)

if skewness > 0:
    print("Interpretation: Right-skewed distribution")
elif skewness < 0:
    print("Interpretation: Left-skewed distribution")
else:
    print("Interpretation: Symmetric distribution")
```

### 5. Coin Toss Simulation

```python
coin_tosses = np.random.choice([0, 1], size=1000)

heads = np.sum(coin_tosses)
estimated_prob = heads / 1000

print("Coin Toss Simulation")
print("Heads count:", heads)
print("Estimated Probability of Heads:", estimated_prob)
print("Theoretical Probability:", 0.5)
```

# Part B - Stretch Problem

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

# Generate samples
normal_data = np.random.normal(loc=50, scale=10, size=1000)   # mean=50, std=10
uniform_data = np.random.uniform(low=20, high=80, size=1000)  # range 20 to 80
```

```python

# Compute statistics
print("Normal Distribution:")
print("Mean:", np.mean(normal_data))
print("Variance:", np.var(normal_data))
print("Standard Deviation:", np.std(normal_data))

print("\nUniform Distribution:")
print("Mean:", np.mean(uniform_data))
print("Variance:", np.var(uniform_data))
print("Standard Deviation:", np.std(uniform_data))
```

```python
# Plot histograms
plt.figure(figsize=(10, 5))

plt.hist(normal_data, bins=30, alpha=0.6, label='Normal Distribution')
plt.hist(uniform_data, bins=30, alpha=0.6, label='Uniform Distribution')

plt.title("Comparison of Normal and Uniform Distributions")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```


### Comparison

Shape

- Normal distribution shows a bell-shaped curve.
- Uniform distribution shows nearly equal frequency across all values (it shows distribution of the form y = constant)

Spread

- Normal distribution has most values concentrated around the mean.
- Uniform distribution spreads values evenly across the interval.

Central tendency

- Normal distribution has strong clustering near the mean.
- Uniform distribution mean lies near the center of the interval, but values do not cluster.

### Real-world use

Normal distribution

Uniform distribution

- Random number generation
