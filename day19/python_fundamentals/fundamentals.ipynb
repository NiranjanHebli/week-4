### 1. Using loops, functions implemented to compute:
- Mean
- Median
- Variance 

```python
# Calculates the mean of a list of numbers using a for loop
def mean_value(numbers):
    """ Calculates the mean of a list of numbers."""
    total = 0
    for num in numbers:
        total += num
    return total / len(numbers)
```

```python
# Calculates the median of a list of numbers using a for loop
def median(numbers):
    """ Calculates the median of a list of numbers."""
    sorted_nums = sorted(numbers)
    n = len(sorted_nums)
    if n % 2 == 0:
        return (sorted_nums[n//2 - 1] + sorted_nums[n//2]) / 2
    else:
        return sorted_nums[n//2]
```

```python
# Calculates the variance of a list of numbers using a for loop
def variance(numbers):
    """ Calculates the variance of a list of numbers."""
    mean = mean_value(numbers)
    total = 0
    for num in numbers:
        total += (num - mean) ** 2
    return total / (len(numbers) - 1) if len(numbers) > 1 else 0
```

```python
data = [10, 20, 30, 40, 50]
print("Mean:", mean_value(data))
print("Median:", median(data))
print("Variance:", variance(data))
```

### 2. Using Higher-order functions (map, filter, reduce where applicable)

```python
from functools import reduce
import math
```

```python
# Calculates the mean of a list of numbers using reduce
def mean_reduce(data):
    """ Calculates the mean of a list of numbers."""
    total = reduce(lambda x, y: x + y, data)
    return total / len(data)
```

```python
# Calculates the median of a list of numbers using sorted and slicing
def median(numbers):
    """ Calculates the median of a list of numbers."""
    sorted_nums = sorted(numbers)
    n = len(sorted_nums)
    if n % 2 == 0:
        mid1, mid2 = sorted_nums[n//2 - 1:n//2 + 1]
        return (mid1 + mid2) / 2
    return sorted_nums[n//2]
```

```python
# Calculates the variance of a list of numbers using map and reduce 
def variance_reduce(numbers):
    """ Calculates the variance of a list of numbers."""
    mean = mean_reduce(numbers)
    squared_diffs = map(lambda x: (x - mean) ** 2, numbers)
    total_squared_diff = reduce(lambda x, y: x + y, squared_diffs)
    return total_squared_diff / (len(numbers) - 1) if len(numbers) > 1 else 0
```

```python
data = [10, 20, 30, 40, 50]
print("Mean (reduce):", mean_reduce(data))
print("Median (sorted):", median(data))
print("Variance (reduce):", variance_reduce(data))
```

### 3. Student marks dataset

```python
marks = [78, 45, 90, 67, 82, 55, 39]

avg = mean_value(marks)

# Students scoring above average
above_avg = list(filter(lambda x: x > avg, marks))

print("Average:", avg)
print("Above Average:", above_avg)

# Grades using list comprehension
grades = ['A' if m >= 80 else 'B' if m >= 50 else 'C' for m in marks]
print("Marks:", marks)
print("Grades:", grades)
```

### 4. Function to compute:  Standard deviation

```python
# Calculates the standard deviation of a list of numbers using the variance calculated with reduce
def std_dev(numbers):
    """ Calculates the standard deviation of a list of numbers."""
    var = variance_reduce(numbers)
    return math.sqrt(var)
```

```python
import numpy as np
```

```python
data = [10, 20, 30, 40, 50]
print("Std Dev:", std_dev(data))
print("Numpy Std Dev:", np.std(data, ddof=1))
```

### 5. Given two sample data, computing mean, variance and difference in their mean

```python
A = [12, 23, 3, 34, 45, 56, 5]
B = [12, 1, 3, 1, 1, 2, 3, 4, 5, 3, 4]

mean_A = mean_reduce(A)
var_A = variance_reduce(A)

mean_B = mean_reduce(B)
var_B = variance_reduce(B)

difference = mean_A - mean_B

print("Group A Mean:", mean_A)
print("Group A Variance:", var_A)

print("Group B Mean:", mean_B)
print("Group B Variance:", var_B)

print("Difference in Means:", difference)
```
