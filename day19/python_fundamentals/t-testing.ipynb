```python
import math
from scipy.stats import t
```

```python
# calculating mean using a simple function
def mean(data):
    """Calculates the mean of a list of numbers."""
    return sum(data) / len(data)
```

```python
# calculating variance using a simple function
def variance(data):
    """ Calculates the variance of a list of numbers."""
    m = mean(data)
    return sum((x - m) ** 2 for x in data) / (len(data) - 1)
```

```python
# calculating two-sample t-test
def two_sample_t_test(sample1, sample2, alpha=0.05):
    """ Performs a two-sample t-test and returns the results."""
    n1, n2 = len(sample1), len(sample2)

    mean1, mean2 = mean(sample1), mean(sample2)
    var1, var2 = variance(sample1), variance(sample2)

    # pooled standard error
    se = math.sqrt((var1 / n1) + (var2 / n2))

    # t-statistic
    t_stat = (mean1 - mean2) / se

    # degrees of freedom
    df = n1 + n2 - 2

    # critical value (two-tailed)
    critical = t.ppf(1 - alpha/2, df)

    # conclusion
    if abs(t_stat) > critical:
        result = "Reject H0: Samples are significantly different"
    else:
        result = "Fail to Reject H0: No significant difference"

    return {
        "mean1": mean1,
        "mean2": mean2,
        "variance1": var1,
        "variance2": var2,
        "t_statistic": t_stat,
        "critical_value": critical,
        "result": result
    }
```

```python
sample1 = [12, 15, 14, 10, 13, 16, 11]
sample2 = [22, 25, 24, 20, 23, 21, 26]

print("Sample 1:", sample1)
print("Sample 2:", sample2)
output = two_sample_t_test(sample1, sample2)

for k, v in output.items():
    print(k, ":", v)
```
